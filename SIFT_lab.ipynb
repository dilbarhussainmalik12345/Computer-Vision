```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'C:\\Users\\Asad Computrs\\Downloads\\opencv-python-tutorials-main\\opencv-python-tutorials-main\\demoImages\\tesla.jpg'
image = cv2.imread(image_path)
```

```python
if image is None:
    print(f"Error: Unable to load image at {image_path}")
    exit(1)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.SIFT_create()
```

```python
keypoints, descriptors = sift.detectAndCompute(gray_image, None)
```

```python
output_image = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('SIFT Keypoints')
plt.axis('off')
plt.show()
```
